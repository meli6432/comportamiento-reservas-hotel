## Librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Archivo

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/reservas_hotel_simulado.csv")

df.head()
```

## Preparación de datos

```python
# Convertir columna de cancelación a formato booleano
df['cancelada_bool'] = df['cancelada'].map({'Sí': 1, 'No': 0})

# Verificar valores nulos
print(df.isnull().sum())
```

## Cancelaciones por canal de reserva

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="canal_reserva", hue="cancelada")
plt.title("Cancelaciones por canal de reserva")
plt.ylabel("Cantidad")
plt.xlabel("Canal de Reserva")
plt.show()
```

El canal con menor cantidad de cancelaciones fue Booking y los canales con mayor cancelación fueron WhatsApp y Web.

## Distribución de noches por reserva

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="noches", bins=7, kde=True)
plt.title("Distribución de noches por reserva")
plt.xlabel("Noches")
plt.ylabel("Frecuencia")
plt.show()
```

La mayoría de las estancias duran entre 5 y 7 noches.

## Distribución por país de origen

```python
plt.figure(figsize=(6, 6))
pais_counts = df['pais_origen'].value_counts()
plt.pie(pais_counts.values, labels=pais_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribución por país de origen")
plt.show()
```

Perú es el país de origen principal, seguido de México y Argentina.

## Cancelación por tipo de habitación

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="tipo_habitacion", hue="cancelada")
plt.title("Cancelaciones por tipo de habitación")
plt.ylabel("Cantidad")
plt.xlabel("Tipo de habitación")
```

Este análisis puede ayudar al hotel a implementar políticas de cancelación diferenciadas por canal
y diseñar estrategias para incentivar reservas con más anticipación.
